```python
import librosa
import os


audio_data_path = "./small_subset/target_directory"
audio_files = [os.path.join(audio_data_path, file).replace('\\', '/') for file in os.listdir(audio_data_path) if file.endswith(".wav")]
print("audio_files=",audio_files)



import numpy as np
import keras
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split
import pandas as pd
import tensorflow as tf
from keras.utils import to_categorical


# Load the tsv file
df = pd.read_csv("./tsv_files/small_tsv_files.tsv", delimiter='\t')

# Modify the audio file path to match the format in the variable "file"
df['path'] = df['path'].apply(lambda x: x.replace('.mp3', '.wav'))

# Create a dictionary with the audio file paths as keys and the transcripts as values
transcript_dict = df.set_index('path')['sentence'].to_dict()

# Extract the transcript for each audio file
# transcripts = []

# Load the audio files and extract the MFCC features
X = []
y1 = []
i=0
# os.listdir('./en_train_full_subset/'):

for file in audio_files:
    base_name = os.path.basename(file)
    i += 1
    y, sr = librosa.load(file)
    S = librosa.power_to_db(librosa.feature.melspectrogram(y, sr=sr, n_mels=40))
    mfcc = librosa.feature.mfcc(S=S, n_mfcc=13)
    X.append(mfcc)
    transcript = transcript_dict.get(base_name)
    y1.append(transcript)
    print("i=",i)

# print("y1=",y1)


# Determine the maximum length of the sequences in X
max_length = max([len(X[i][0]) for i in range(len(X))])

X_copy = np.empty((10,13,max_length))

for i in range(len(X)):
    for j in range(len(X[i])):
        print(X[i][j])
        print(max_length - len(X[i][j]))
        print(np.pad(X[i][j], (0, max_length - len(X[i][j]))))
        X_copy[i][j] = np.pad(X[i][j], (0, max_length - len(X[i][j])))



# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_copy, y1, test_size=0.2)

# Reshape the data for the LSTM input
X_train = np.array(X_train).reshape((-1, 13, max_length))
X_test = np.array(X_test).reshape((-1, 13, max_length))



# Define the LSTM model
# transcript_vocab = set(["yes", "no", "one", "two", "three", "four"])


transcript_vocab = set(y1)

num_classes = len(transcript_vocab)
y_train = [list(transcript_vocab).index(x) for x in y_train]
y_train = to_categorical(y_train, num_classes)
y_test = [list(transcript_vocab).index(x) for x in y_test]
y_test = to_categorical(y_test, num_classes)

model = Sequential()
model.add(LSTM(128, input_shape=(13, max_length)))
# model.add(Dense(len(transcript_vocab), activation='softmax'))
model.add(Dense(num_classes, activation='softmax'))

# # Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# # Fit the model to the training data
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

#Tester sur un autre jupyter notebook et corriger, comprendre le code

```
